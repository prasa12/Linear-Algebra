```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 60, 100)
```

```python
mark_1 = 1*t
```

```python
mark_2 = 4*(t-30)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([0,60])
ax.set_ylim([0, 60])
ax.plot(t,mark_1,c='green')
ax.plot(t,mark_2,c='blue')

    
```
