```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 40)
```

```python
t
```

```python
d_r = 2.5 * t
```

```python
d_s = 3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank robber caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,d_r,c='green')
ax.plot(t,d_s,c='blue')
```
